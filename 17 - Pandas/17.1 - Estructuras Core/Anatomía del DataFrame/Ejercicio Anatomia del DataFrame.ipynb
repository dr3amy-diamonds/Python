## Solución de Misión 1: La Construcción

```python
import pandas as pd
import numpy as np

#Crear los datos en el inventario
data_inventario = {
    'Producto': ['Laptop', 'Mouse','Teclado','Monitor'],
    'Precio': [1200, 25, 45, 300],
    'Stock': [5, 50, 20, 10],
    'Disponible':[True, True, True, False]
}

#Crear Dataframe
df = pd.DataFrame(data_inventario)

print("--- Mi Primer DataFrame ---")
print(df)

```

## Solución de Misión 2: El Informe de Dimensiones (Atributos)

```python
print("\n--- Información ---")
print("1. Índice (Axis 0):\n", df.index)
print("2. Variables que se estan midiendo:\n", df.columns)
print("Total de variables:", len(df.columns))
print("3. Valores (NumPy):\n", df.values)
print("4. Forma (Filas, Cols):\n", df.shape)
total_stock = df['Stock'].sum()
print("5. Total de unidades en inventario:\n", total_stock)
```

## Solución de Misión 3: La Prueba de ADN (Heterogeneidad)

```python
#Verificar tipo de dato del dataframe
print("\n--- Tipo de Datos ---")
print("Tipo de dato del dataframe:\n",df.dtypes)
```
