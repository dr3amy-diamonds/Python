## Solución 1: Conversión de Divisas (Broadcasting)

```python
import pandas as pd
import numpy as np
```

```python
#Creación de la Serie
precios_usd=pd.Series([250, 25, 40, 60], index=['Monitor', 'Mouse', 'Teclado', 'Auriculares'])

precios_cop=precios_usd*4050
print("--- Precios convertidos (Broadcasting) ---")
print(precios_cop)
```

# Solución 2: Fusión de Inventarios (Alineación de Índices)

```python
#Creación de las series
bodega_a=pd.Series([10, 5, 20], index=['Laptop', 'Tablet', 'Móvil'])
bodega_b=pd.Series([15, 30, 10], index=['Laptop', 'Reloj', 'Móvil'])

stock_total=bodega_a+bodega_b
print("\n--- Total (Alineado por Índice) ---")
print(stock_total)
```

```python
#Rellenar con (0.0) valores con NaN
relleno=stock_total.fillna(0.0)

print(f"\n--- Serie Corregida (Relleno con (0,0)) ---")
print(relleno)
```

# Solución 3: Datos de Sensores (Manejo de Nulos)


```python
#Crear serie
sensor_amb=pd.Series([22.5, np.nan, 23.0, 21.8, np.nan, 22.2])

# Contar valores nulos
nulos = sensor_amb.isnull().sum()
print(f"\nCantidad de valores nulos: {nulos}")

#Calcular promedio
promedio=sensor_amb.mean()
valores_corregida=sensor_amb.fillna(promedio)

print(f"\n--- Serie Corregida (Relleno con media: {promedio}) ---")
print(valores_corregida)
```

# Solución 4: Clasificación de Estudiantes (Método Apply)

```python
notas_estudiantes=pd.Series([4.8, 2.5, 3.0, 3.9], index=['Ana', 'Carlos', 'Luis', 'Maria'])

def sistema_notas(notas):
    if notas >= 3.0:
        return "Aprobado"
    else:
        return "Reprobado"

#Aplicamos esa función a cada datos de la serie
estado_notas=notas_estudiantes.apply(sistema_notas)

print("--- Transformación con .apply() ---")
print(estado_notas)
```
