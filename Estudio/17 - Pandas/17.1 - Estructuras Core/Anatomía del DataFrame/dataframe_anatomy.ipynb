# 1. Creación: De Diccionario a DataFrame
La teoría decía: "Un DataFrame es un diccionario de listas organizadas"..

```python
import pandas as pd

# 1. Los datos crudos (Diccionario)
datos = {
    'Producto': ['Laptop', 'Mouse', 'Monitor'],
    'Precio': [1000, 20, 200],
    'Stock': [10, 50, 15]
}

# 2. La magia (Constructor)
df = pd.DataFrame(datos)

print("--- Mi Primer DataFrame ---")
print(df)
```

## 2. Anatomía Interna: Inspeccionando las partes
La teoría decía: "Tiene 3 componentes accesibles: Index, Columns, Values".

```python
print("\n--- Disección ---")
print("1. Índice (Axis 0):\n", df.index)
print("2. Columnas (Axis 1):\n", df.columns)
print("3. Valores (NumPy):\n", df.values)
print("4. Forma (Filas, Cols):\n", df.shape)
```

Si ejecutas esto, verás que .values elimina los nombres y te muestra solo la matriz de datos.

## 3. La Prueba de ADN: Extracción de Series
La teoría decía: "Una columna aislada es una Serie".

```python
#Extraemos una sola columnas
columna_precio=df['Precio']

print("\n --- Prueba de ADN ---")
print("¿Qué es esto?:",type(columna_precio))
print(columna_precio)
```

El resultado de type() te confirmará que, efectivamente, es una pandas.core.series.Series.
