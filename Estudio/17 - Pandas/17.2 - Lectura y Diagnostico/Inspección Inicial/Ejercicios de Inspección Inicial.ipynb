# El Sesgo del Orden (.sample vs .head)

```python
word_list=['Correcta']*20

#Convertir a DataFrame
df=pd.DataFrame(word_list)

#Cambiar elemento accediendo mendiante el index
df.loc[14] = "ERROR CRÍTICO"

print("--- Inspeccionar rápidamente ---")
print("Usando .head()", df.head(5))
print("Usando .sample()",df.sample(10))
```

# Conclusión
Validar datos utilizando únicamente el método .head() no es suficiente porque este solo permite observar las primeras filas del conjunto de datos, lo que puede generar una falsa sensación de que la información es correcta y homogénea. Como se evidencia en el ejemplo, aunque las primeras observaciones contienen valores aparentemente válidos, existe un error crítico ubicado en una posición intermedia del DataFrame que .head() no logra detectar. Por ello, es necesario complementar la inspección con otros métodos como .sample() o revisiones más exhaustivas, que permitan analizar datos en distintas posiciones y así identificar inconsistencias, errores atípicos o valores inesperados que podrían afectar el análisis y la toma de decisiones.

# Solución 4: El Mapa de Calor de Nulos (.isna)
