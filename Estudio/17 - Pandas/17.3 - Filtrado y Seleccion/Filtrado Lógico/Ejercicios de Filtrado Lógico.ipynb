# Solución 1: El Portero de Discoteca (Filtro Simple)

```python
data_club= pd.DataFrame({
    "Nombre": ["Juan", "Sofia", "Mateo", "Lucia", "Pedro"],
    "Edad": [15, 19, 25, 17, 18]
})

print(data_club)
```

```python
#Crear máscara edad >18
mask_edad = data_club['Edad'] >= 18
print(mask_edad)

#Usar la máscara para filtrar
df_admitidos = data_club[mask_edad]

print(df_admitidos)
```

# Solución 2: Recursos Humanos (Lógica AND)

```python
#Creacion de datos y dataframe
df_rrhh= pd.DataFrame({
    "Candidato": ["Ana", "Luis", "Marta", "Javi"],
    "Lenguaje": ["Python", "Java", "Python", "Python"],
    "Experiencia": [2, 5, 4, 1]
})

#Crear condicional
check_candidate=(df_rrhh['Lenguaje']=='Python') & (df_rrhh['Experiencia']>3)
print("------CANDIDATO(s) que cumple con esta condición")
df_candidate = df_rrhh[check_candidate]
print(df_candidate)
```

# Solución 3: El Bibliotecario Eficiente (`isin` y `between`)

```python
#Creación del dataframe
df_biblioteca = pd.DataFrame({
    "Título": ["Libro A", "Libro B", "Libro C", "Libro D", "Libro E"],
    "Año": [1995, 2005, 1998, 1992, 2020],
    "Género": ["Terror", "Romance", "Ciencia Ficcion", "Terror", "Historia"]
})

#Creación de las mascaras(También de filtro)
df_filtro=(df_biblioteca['Año'].between(1990, 1999)) & (df_biblioteca['Género'].isin(['Terror', 'Ciencia Ficcion']))
print("--------LIBROS----------")
df_books=df_biblioteca[df_filtro]
print(df_books)
```

# Solución 4: Control de Calidad (`~ NOT` y `.query()`)

```python
productos = pd.DataFrame({
    "Producto": ["Tornillo", "Tuerca", "Clavo", "Arandela"],
    "Estado": ["OK", "Defectuoso", "OK", "Pendiente"],
    "Peso": [10, 45, 12, 30]
})

#Con ~ NOT
mask = (~(productos['Estado'] == 'OK')) & (productos['Peso'] < 50)
print("---RESULTADO ~ NOT")
df_resul = productos[mask]
print(df_resul)

# Con .query()
df_query = productos.query("Estado != 'OK' and Peso< 50")
print("RESULTADOS .query()")
print(df_query)
```
