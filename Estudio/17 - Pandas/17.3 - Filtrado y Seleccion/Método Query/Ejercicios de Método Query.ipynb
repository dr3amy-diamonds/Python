```python
import pandas as pd

# PREPARACIÓN DEL LABORATORIO
data = {
    'nombre': ['Ana', 'Luis', 'Carlos', 'Sara', 'Marta', 'Pedro', 'Julia'],
    'departamento': ['Ventas', 'IT', 'Ventas', 'HR', 'IT', 'Ventas', 'HR'],
    'salario': [5000, 8000, 4500, 6000, 8200, 4800, 7500],
    'años_exp': [2, 5, 1, 4, 6, 2, 5],
    'activo': [True, True, False, True, True, False, True]
}
df_data = pd.DataFrame(data)

#Hay algunos cambios de variables como lo pide el ejercicio, pero no se han hecho cambios en los nombres de las columnas para que el código sea más fácil de entender.
```

# Solución 1: El Gerente de TI (Filtro Simple)

```python
df_it=df_data.query("departamento=='IT'")
print("Empleados del departamento de IT:")
print(df_it)
```

# Solución 2: Cazatalentos (Lógica AND)

```python
df_perfecto=df_data.query("departamento == 'IT' and años_exp >4")
print(("Candidatos perfectos para el departamento de IT con más de 4 años de experiencia:"))
print(df_perfecto)
```

# Solución 3:Auditoría de Anomalías (Lógica OR)

```python
df_auditar=df_data.query("salario > 7000 or años_exp < 2")
print("Empleados con salario mayor a 7000 o con menos de 2 años de experiencia:")
print(df_auditar)
```

# Solución 4: El Presupuesto Dinámico (Variables con @)

```python
#Creación de la variable dinámica
limite_bono=6000
df_bono=df_data.query("salario < @limite_bono")
print(f"Empleados con salario menor a {limite_bono} que podrían ser elegibles para un bono:")
print(df_bono)
```
